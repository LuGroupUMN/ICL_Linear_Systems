```python
import matplotlib.pyplot as plt
import numpy as np

### Plots for diversity l2
npy_name = 'RM_Result_diversity_test.npy'
loaded_items = []
with open(npy_name, 'rb') as f:
    while True:
        try:
            item = np.load(f)
            loaded_items.append(item)
        except EOFError:
            break
m_ls, para_ls, op_l2_mat, bad_l2_mat = loaded_items

plt.rcParams.update({
    'font.size': 14,          # base font size
    'axes.titlesize': 16,     # subplot title
    'axes.labelsize': 14,     # x and y labels
    'xtick.labelsize': 12,    # x tick labels
    'ytick.labelsize': 12,    # y tick labels
    'legend.fontsize': 12,    # legend
    'figure.titlesize': 18    # figure title (if using suptitle)
})


fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].loglog(m_ls, op_l2_mat[0, ...], label=r'$V(x)=c \mathbf{I}, c \sim U(1,2)$')
for i in range(1, len(para_ls)):
    axs[0].loglog(m_ls, op_l2_mat[i, ...], '--*', label=rf'$V(x) \sim \mathbf{{U}}({str(para_ls[i][0])},{str(para_ls[i][1])})$')
axs[0].set_xlabel(r'$m$', fontsize=18)
axs[0].legend(fontsize=11, loc='lower left')

axs[1].loglog(m_ls, bad_l2_mat[0, ...], label=r'$V(x)=c \mathbf{I}, c \sim U(1,2)$')
for i in range(1, len(para_ls)):
    axs[1].loglog(m_ls, bad_l2_mat[i, ...], '--*', label=rf'$V(x) \sim \mathbf{{U}}({str(para_ls[i][0])},{str(para_ls[i][1])})$')
axs[1].set_xlabel(r'$m$', fontsize=18)
axs[1].legend(fontsize=11, loc='lower left')
plt.tight_layout()
plt.savefig(f'RM_diversity_l2.png', bbox_inches = 'tight')
plt.show()
```

```python
### plots for OOD
import matplotlib.pyplot as plt
import numpy as np

### Load V
npy_name = 'RM_Result_OOD_test_V.npy'
loaded_items = []
with open(npy_name, 'rb') as f:
    while True:
        try:
            item = np.load(f)
            loaded_items.append(item)
        except EOFError:
            break
m_ls, para_V, V_l2_mat = loaded_items

### Load f
npy_name = 'RM_Result_OOD_test_f.npy'
loaded_items = []
with open(npy_name, 'rb') as f:
    while True:
        try:
            item = np.load(f)
            loaded_items.append(item)
        except EOFError:
            break
m_ls, para_f, f_l2_mat = loaded_items

plt.rcParams.update({
    'font.size': 14,          # base font size
    'axes.titlesize': 16,     # subplot title
    'axes.labelsize': 14,     # x and y labels
    'xtick.labelsize': 12,    # x tick labels
    'ytick.labelsize': 12,    # y tick labels
    'legend.fontsize': 12,    # legend
    'figure.titlesize': 18    # figure title (if using suptitle)
})


fig, axs = plt.subplots(1, 2, figsize=(10, 4))
### for V
axs[0].loglog(m_ls, V_l2_mat[0, ...], label=rf'$(a,b)=$(' + str(para_V[0][0])+','+str(para_V[0][1])+')')
for i in range(1, len(para_V)):
    axs[0].loglog(m_ls, V_l2_mat[i, ...], '--*', label=rf'$(a,b)=$(' + str(para_V[i][0])+','+str(para_V[i][1])+')')

axs[0].set_xlabel(r'$m$', fontsize=18)
#axs[0].set_title(r'Shift in $V(x)$', fontsize=18)
axs[0].legend(fontsize=12, loc='lower left')

### for f
axs[1].loglog(m_ls, f_l2_mat[0, ...], label=rf'$\rho={para_f[0]}$')
for i in range(1, len(para_f)):
    axs[1].loglog(m_ls, f_l2_mat[i, ...], '--*', label=rf'$\rho={para_f[i]}$')

axs[1].set_xlabel(r'$m$', fontsize=18)
axs[1].legend(fontsize=12, loc='upper center',
    bbox_to_anchor=(0.5, 0.58),
    ncol=2)
#axs[1].set_title(r'Shift in $f(x)$', fontsize=18)
plt.tight_layout()
plt.savefig(f'RM_OOD_l2.png', bbox_inches = 'tight')
plt.show()
```

```python
### plots for IND
import matplotlib.pyplot as plt
import numpy as np

### Load sn
npy_name = 'RM_Result_test_sn.npy'
loaded_items = []
with open(npy_name, 'rb') as f:
    while True:
        try:
            item = np.load(f)
            loaded_items.append(item)
        except EOFError:
            break
n_n_ls, n_m_ls, n_mse_mat, n_mse_ref_rate = loaded_items

### Load m
npy_name = 'RM_Result_test_m.npy'
loaded_items = []
with open(npy_name, 'rb') as f:
    while True:
        try:
            item = np.load(f)
            loaded_items.append(item)
        except EOFError:
            break
m_m_ls, m_n_ls, m_mse_mat, m_mse_ref_rate,  = loaded_items

### Load N
npy_name = 'RM_Result_test_N.npy'
loaded_items = []
with open(npy_name, 'rb') as f:
    while True:
        try:
            item = np.load(f)
            loaded_items.append(item)
        except EOFError:
            break
N_N_ls, N_m_ls, N_mse_mat, N_mse_ref_rate, = loaded_items



plt.rcParams.update({
    'font.size': 14,          # base font size
    'axes.titlesize': 16,     # subplot title
    'axes.labelsize': 14,     # x and y labels
    'xtick.labelsize': 12,    # x tick labels
    'ytick.labelsize': 12,    # y tick labels
    'legend.fontsize': 12,    # legend
    'figure.titlesize': 18    # figure title (if using suptitle)
})


fig, axs = plt.subplots(1, 3, figsize=(15, 4))

### for n
for i in range(len(n_m_ls)):
    axs[0].loglog(n_n_ls, n_mse_mat[i, ...], '--*', label=f'm={n_m_ls[i]}')

axs[0].loglog(n_n_ls, n_mse_ref_rate, 'b', label=r'$O(n^{-2})$')
axs[0].set_xlabel(r'$n$', fontsize=18)
axs[0].legend(fontsize=12, loc='lower left')

### for m
for i in range(len(m_n_ls)):
    axs[1].loglog(m_m_ls, m_mse_mat[i, ...], '--*', label=f'n={m_n_ls[i]}')

axs[1].loglog(m_m_ls, m_mse_ref_rate, 'b', label=r'$O(m^{-1})$')
axs[1].set_xlabel(r'$m$', fontsize=18)
axs[1].legend(fontsize=12, loc='lower left')

### for N
for i in range(len(N_m_ls)):
    axs[2].loglog(N_N_ls, N_mse_mat[i, ...], '--*', label=f'm={N_m_ls[i]}')

axs[2].loglog(N_N_ls, 1.5*N_mse_ref_rate, 'b', label=r'$O(N^{-1})$')
axs[2].set_xlabel(r'$N$', fontsize=18)
axs[2].legend(fontsize=12, loc='lower left')

plt.tight_layout()
plt.savefig(f'RM_IND_l2.png', bbox_inches = 'tight')
plt.show()
```
